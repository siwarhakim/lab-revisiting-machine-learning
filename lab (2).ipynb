```python
categorical = categorical.drop(drop_list, axis=1)
distinct_value_counts = {}

# Iterate over each column in the DataFrame
for column in categorical.columns:
    # Calculate the distinct value counts for the column
    distinct_values = categorical[column].value_counts()
    # Store the distinct value counts in the dictionary
    distinct_value_counts[column] = len(distinct_values)

# Print the distinct value counts for each variable
for variable, count in distinct_value_counts.items():
    print(f"Variable '{variable}' has {count} distinct values.")
```
